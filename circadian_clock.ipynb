# the circadian clock

The daily schedules of animals, plants, and even bacteria are controlled by an internal timekeeper called the circadian clock. Anyone who has experienced the misery of jet lag knows that this clock never stops ticking. Rats and research volunteers alike, when placed in a bunker, naturally maintain a roughly 24-hour cycle of activity and rest in total darkness. And, like any timepiece, the circadian clock can malfunction, resulting in a genetic disease known as delayed sleep-phase syndrome (DSPS).

The circadian clock must have some basis on the molecular level, which presents many questions. How do individual cells in animals and plants (let alone bacteria) know what time it is? Is there a â€œclock geneâ€? Can we explain why heart attacks occur more often in the morning, while asthma attacks are more common at night? And can we identify genes that are responsible for â€œbreakingâ€ the circadian clock to cause DSPS?

In the early 1970s, Ron Konopka and Seymour Benzer identified mutant flies with abnormal circadian patterns and traced the fliesâ€™ mutations to a single gene. Biologists needed two more decades to discover a similar clock gene in mammals, which was just the first piece of the puzzle. Today, many more circadian genes have been discovered; these genes, having names like timeless, clock, and cycle, orchestrate the behavior of hundreds of other genes and display a high degree of evolutionary conservation across species.

We will first focus on plants, since maintaining the circadian clock in plants is a matter of life and death. Consider how many plant functions depend on when the sun rises and sets. Indeed, biologists estimate that over a thousand plant genes are circadian, including the genes related to photosynthesis, photo reception, and flowering. But what does it mean for a gene be circadian?



# Gene expression


The Central Dogma of Molecular Biology states that â€œDNA makes RNA makes protein.â€ The DNA corresponding to a gene is first transcribed into a strand of RNA composed of four ribonucleotides: adenine, cytosine, guanine, and uracil (which replaces thymine in DNA). Then, the RNA transcript is translated into the amino acid sequence of a protein, which performs some function in the cell.

Much like DNA replication, the chemical machinery underlying transcription and translation is fascinating, but from a computational perspective, both processes are straightforward. Transcription simply replaces all occurrences of T in a DNA string with U. The resulting strand of RNA is translated into an amino acid sequence as follows. During translation, the RNA strand is partitioned into non-overlapping 3-mers called codons. Then, each codon is converted into one of 20 amino acids via the genetic code; the resulting sequence can be represented as an amino acid string over a 20-letter alphabet. As illustrated in the figure below, each of the 64 RNA codons encodes its own amino acid (some codons encode the same amino acid), with the exception of three stop codons that do not translate into amino acids and serve to halt translation (see DETOUR: Discovery of Codons and Split Genes). For example, the DNA string "TATACGAAA" transcribes into the RNA string "UAUACGAAA", which in turn translates into the amino acid string "YTK". 

The key point is that cells are able to transcribe different genes into RNA at different rates. This variance in the production of a geneâ€™s transcripts, or gene expression, explains how a brain cell and a skin cell can have the same DNA but perform vastly different functions. Variation in gene expression through the day also accounts for how the cell can keep track of time.

# Regulatory proteins

It turns out that every plant cell keeps track of day and night independently of other cells, and that just three plant genes, called LHY, CCA1, and TOC1, are the clockâ€™s master timekeepers. Such genes, and the regulatory proteins that they encode, are often controlled by external factors (e.g., nutrient availability or sunlight) in order to allow organisms to adjust their gene expression.

For example, regulatory proteins controlling the circadian clock in plants coordinate circadian activity as follows. TOC1 promotes the expression of LHY and CCA1, whereas LHY and CCA1 repress the expression of TOC1, resulting in a negative feedback loop. In the morning, sunlight activates the transcription of LHY and CCA1, triggering the repression of TOC1 transcription. As light diminishes, so does the production of LHY and CCA1, which in turn do not repress TOC1 any more. Transcription of TOC1 peaks at night and starts promoting the transcription of LHY and CCA1, which in turn repress the transcription of TOC1, and the cycle begins again.

LHY, CCA1, and TOC1 are able to control the transcription of other genes because the regulatory proteins that they encode are transcription factors, or master regulatory proteins that turn other genes on and off. A transcription factor regulates a gene by binding to a specific short DNA interval called a regulatory motif, or transcription factor binding site, in the geneâ€™s upstream region, a 600-1000 nucleotide-long region preceding the start of the gene. For example, CCA1 binds to "AAAAAATCT" in the upstream region of many genes regulated by CCA1.

# ðŸ§¬ Motifs in Molecular Biology

## What are Motifs?

In the context of molecular biology, **motifs** are short, recurring patterns in DNA that have a **biological function**. These patterns are usually **recognized by proteins**, particularly **transcription factors**.

---

## Function of Motifs

- âœ¨ **Motifs serve as binding sites** for **transcription factors** (TFs), which are regulatory proteins.
- ðŸ§  The presence of a motif in a DNA sequence indicates a **region that may regulate gene expression**.
- ðŸ§¬ These motifs often appear in **promoter regions** or **enhancers**, helping control when and where a gene is turned on or off.

---

## Are Motifs Proteins?

âŒ No, motifs are **not proteins**.

âœ… They are **specific DNA sequences** that **proteins bind to**.

---

## Transcription Factors and Their Role

- Transcription factors are **proteins** that regulate transcription.
- They **bind to motifs** in the DNA sequence to influence the activity of RNA polymerase.
- Example: A TF may bind to a **TATA box motif** to initiate transcription.

---

## Quick Summary

| Component             | Type       | Function                                    |
|----------------------|------------|---------------------------------------------|
| Motif                | DNA sequence | Acts as a binding site                     |
| Transcription Factor | Protein     | Binds to motif to regulate gene expression |

---

## Examples of DNA Motifs

- **TATA box**: A common promoter motif.
- **CAAT box**: Found upstream of many eukaryotic genes.
- **GC-rich regions**: May attract different types of transcription factors.

```python
def profile_matrix_with_pseudocounts(motifs):
    k = len(motifs[0])
    profile = {'A': [1] * k, 'C': [1] * k, 'G': [1] * k, 'T': [1] * k}
    total_motifs = len(motifs) + 4  # pseudocounts for each nucleotide

    for i in range(k):
        for motif in motifs:
            profile[motif[i]][i] += 1

    for nucleotide in profile:
        for i in range(k):
            profile[nucleotide][i] /= total_motifs

    return profile


def profile_most_probable_kmer(text, k, profile):
    max_prob = -1
    most_probable_kmer = text[:k]

    for i in range(len(text) - k + 1):
        kmer = text[i:i + k]
        prob = 1
        for j in range(k):
            prob *= profile[kmer[j]][j]
        if prob > max_prob:
            max_prob = prob
            most_probable_kmer = kmer

    return most_probable_kmer


def greedy_motif_search_with_pseudocounts(dna, k, t):
    best_motifs = [string[:k] for string in dna]

    for i in range(len(dna[0]) - k + 1):
        motifs = [dna[0][i:i + k]]

        for j in range(1, t):
            profile = profile_matrix_with_pseudocounts(motifs)
            motifs.append(profile_most_probable_kmer(dna[j], k, profile))

        if score_motifs(motifs) < score_motifs(best_motifs):
            best_motifs = motifs

    return best_motifs


def score_motifs(motifs):
    score = 0
    k = len(motifs[0])

    for i in range(k):
        column = [motif[i] for motif in motifs]
        most_common = max(set(column), key=column.count)
        score += sum(1 for nucleotide in column if nucleotide != most_common)

    return score


# Sample Input
dna = [
"TGTGCCGCTTCCACACAATTTACCAAGGGCCCCAGAGTTAGACGCTATGTGCCGATTACCGTAAGCCATGAGTGTGTTACGGTTGGTAGGTGAGATTTCGTAGTCAGTCTCACGCACCTATGTGTTTCTTGCACCCCGTAGGTGGGGGACGTCCGCC",
"CTCCTCGTAGTGCGGGTCGTCCGCGCAGCACCCGGCTAAGTATATGCGTGCGGCACAGTTCGGATCAGTTGCTCGCGGTATCCTGTATCTACGACGGCGGGGCCACGGGATCACTATCGAGTGCGTACCCAGACGCGATGGGCATGTTTGCTATTC",
"TGGCGGCATCGAATTGTATAGATCCATGGATTCTTCATTTCCTGGAGGGTGTCTTTTATCTTAGATGCCCTCGAGAATTAAAAGGCCCACCCCTTGCTAGTGTGCGCAGGCTTATTTCTGGAATCTAACACCCGAAGTGACTGTTGGTTCGTCCGC",
"TGAGGCGTTTAGATCCTTATAACTTGGAACGTCCCCTAACTGTCCGACGGACCTTATCGGGCTCTATGATATCTATCCGGACCTGGAGCGTGGAGGAATCCTAAAGGCCGTTACCAACGATCTCCTTTTTATTACGATGGCGATTGCTAACAAATG",
"AGGGCACAGTTAGGACCGCACTGAGTTTACTGCATTGGTATTGCTGTCGTATCCGGCGTCTGACAACTAGCCATTTCCCTAACAAAGCGACCCCAGCGAGCGGAACGCCTAGTCCTGCCTTCAACTCAGCAGCGTTTGATAGGGGGGACCGTCCTC",
"CTCGAGCAAAACTGAACCCGGATAGGGCTCGTCCTCGTTAACATAAAGCAGCATACCAGAGCCCCACTACGATCTAAATACTTAATCAGAACGATTTTGTTTTGCGTAGGCACCCAGTTCGACCAACCAGGCCGCCTTGGTGGACGGCTAGGATTG",
"CGGTGTGGTCTTCAGCATTTACATCCTTGTAATAATCGGTAACAGGCCAAATCGTGAAAGCAGGCCGAAAACCCCACTGGCGCTCGGGCCGTCCCCATACACACTAAATAATGAAACTGTGCGCTTAGATGTTGTCAATAACCGCGCGGGGCAATC",
"GCGTTCCGACCGCTGAGTTACTAGCTAATTGGCCTGACTGACCAACGGATTAAACGTAGAAGTACGGAATATGTGCCGCCAAGCAAGGACGGTAGCTTCAAATAAAAAAGGCGAAGCTAATGGGGCGTCCCCTCGCGCCCCCTGAGTATCTACTAG",
"TTTTATGTGCTTCGGAGCTGTGTTTCGATTCCCGGTTTAGCTAGATACCTCGCAATTCTGAGTGTGTCTCGCTGGCCCGTCCCCTAAACTTCCGCGTACCTAGTGCGCGCGGTGTAGGAACATGGAATTTCCAAATTCATAGCAGGGGCTCAATCT",
"GTGTGCCATCCACTTCTGATGTACGGGCAGGCGGCTAGAGATTTTTGCGTCTGAAGTAATCCAATCTCCTGCGTCGTAATGGCCAGGTGCGTCCTCCAGATCAGCCGCACTGAGATACCGTCAAGTCGACTAGCGAGCCCGCATGGCATTTGAAGA",
"CATTACCCATTAACTAAGCATCTTTGTACGATTCTCCCCGTTTTTGTTGTCAAGGGCCTGCAATCCACATCTAGGGACGTCCGCAGGGTACACTCCGTCAACCAGTAGGGAGACTTCAGTCTACTACTCGACTTTGACGCGACGTGGCCGAGATAA",
"GGGAACGTCCACAGTCGCCCAGACTCCCTGCATTAGATACAAATAACGATTCATCTGCTCGCCTACCTTTTACAGTCTTCGTATAAAAGCTAACCGGAAGCGCGACAGTGAGAACGGGAAGCAAGATGACGAACCTATGAGAAGACGTGCTGGGAT",
"AACCGCTCAGATAGGGCCGTCCACTTCACTCATACGAGCTTCTTCTCGGCGGAGTGGGTCAGGGGTCACTTAATTTTAGGCACCAAAGGACTGCCGTTAATCTCCGATACGTCGGTAGACATATAAATCGAAGCGCAAATCTTAGCTGATCCTTGG",
"GGGGGGTCGTCCTCTTACCGAATCCTAAAAGCGGCCTTTAGTCGAGAGGTGCAAGAATCCTTTTGCAACTAGATCATCGTGAGGGATATGACACGAACTCTTTGAGAGCTAACCCCGGTCACCTTGCCATACTCGTCTCGGACTAGCCCAAATATAGG",
"AGGCTCGTCCGCCCTAGTGGGAGCCAGTTGTAATAGTGGGGCCCTCCACTTGCGCGTCAGCACTTACGACCATGACGCTGAGGGTTGGAGTGGTGACGATAAACACTAACAGGCTCATAGGAATGAGTGTATAGTGTCCTTTCGTGCGCAAGCGGT",
"GTTTTCCTACCCATAATCTTTAAATACCATCATCTACGTGTGGATGCTATTGGAGTACCGTGGTACGTCCGCTAGTTCAAAATGGCCAACAGGGCGCACATTCAATAACTATTAATATGTGCGTTTGAAGACAAGCGGCTAAGGGAAAACAAAGAT",
"TTTCGTTCCACCTTTGTGTGCCTATCGTCTAAATTAAGCGGGTTTACGCGCGGTGCCATAACGATGCGAACTAATGGAGTTGCATGTCCTGATACGTGGTACGTCCGCATTGCCCGATGTATGGGTACGACAAGCGTATCGCCTACATTATCCACC",
"TGGAGCGTCCCCTAATTCTGAGTGATAGCTTTCCTTAAGACATTCTACGCCACTCAAGCGACAACAAATTAGTACGCCGTAACGCTTATTTGTGGGTTGCCTTTTCCGCCTCGCAAAATAATCTGCAACGACTGGATGATTGTTAATATCAAGCGT",
"GATCGATTCCCTAGAATATAGCGTTATTCTGTTAACCATTACTCGACATGCAGTCATTAACCAGTATGGTGACCCGCTTTAGACATTTAAGTCCCAAGTCTAACCCGATGGGCCGTCCGCCACGCTTACGACACGTAGACCATAGGTGGGGTAGTT",
"ATTCACAAGAAATGGTACGTCCACTACGCTATGACGGGGCATTGCGCCAGACTAAGGACCCTCCTCCGATCGAACAAATTCTAGTAGGCTACGATCATTCAGTCGTCTCAGACTAAATTTCCAACGATATTAAGATTCAATCGTTGATCGACTCAA",
"ACAGTCCGTTGCGGGGCCGTCCGCGCCTGTGCACCAAGCTCCTCTTTTCCCAAATAAGGATGTCGTCTTTCTAATTCAAGTGTTACGAAACCTATAAATGGCTGCCAGGAGCTACATGTCTCACAAACGCGTGGGGTGTTCGTTTCCCACACTAGC",
"CGACACACGAAGTGGATGATGTTCTGCCACTTGATCATCCCATAAGCTAGAAGGTAGCCCTTGTCCTCCCCAACGCTCTCTGCCCGATCACTCATTTAGAAAACGTCCCGCCTGTCTGATCTGTGGCAGACCGGGGCCGTCCTCAGCGGTGGTCCG",
"CTGCAATCCATGCCGCGTGTAGGGGGTCGTGTTGTCTGGCGCGTCCACACAAAAGTGGTTATGCCGGGTGGCTCGGGCCCTGAGAATCTAAAGCCCAGAGTTATCCACTCCGCCTGCGCGATCGTTAGTTCGGCATAGGCCATAGTCCGAGTGAGC",
"CCGTAGCTTCGCCTGTGAAGACATTGGTAATCTAAATGCGCGGTAGAGCGAGGGTTTGTAAGATATGTGCAGGTGGCCTCATACGCTCACTTTTATGTGTACACAGATAAGCCCCGAGAGGGGATCGTCCGCGATTAAATTTGCCTCCATAGATCA",
"ACAAAGTTGATAATGACGATTCGAACACTGGGACTGATATTAAGATCTTCTGTCAGGACCAGTGACGTGCAATAATTTTTTCCACGGGGCGTCCCCCGTCTCGACGTGCATGACCGACAGAGCATGATTCGTATCGAGCCTACAGTAGCAACAAAG"
]
k = 12
t = 25

# Sample Output
result = greedy_motif_search_with_pseudocounts(dna, k, t)
print(" ".join(result))
print(*result,sep = " ")
```
