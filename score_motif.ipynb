# SCORE MOTIF

we can compute Score(Motifs) by first constructing Consensus(Motifs) and then summing the number of symbols in the j-th column of Motifs that do not match the symbol in position j of the consensus string.

we basicly have to keep adding the number of nucleotides gathered from the **j-th column** which is not the part of **CONSENSUS**

- Sample Input:
```python
AACGTA
CCCGTT
CACCTT
GGATTA
TTCCGG
```
- Sample Output:

*14*

````python
def Concensus(Motifs):
    k = len(Motifs[0])
    count = Count(Motifs)
    concensus = ""
    for j in range(k):
        m = 0
        frequentsymbol = ""
        for symbol in "ATGC" :
            if count[symbol][j] > m :
                m = count[symbol][j]
                frequentsymbol = symbol
        concensus = concensus + frequentsymbol
    return concensus
# Copy your Count(Motifs) function here.
def Count(Motifs):
    count = {}
    k = len(Motifs[0])
    for symbol in "ATGC":
        count[symbol] = []
        for j in range(k):
            count[symbol].append(0)
    t = len(Motifs)
    for i in range(t):
        for j in range(k):
            symbol = Motifs[i][j]
            count[symbol][j] += 1
    return count
# Input:  A set of k-mers Motifs
# Output: The score of these k-mers.
def Score(Motifs):
    t = len(Motifs)
    k = len(Motifs[0])
    concensus = Concensus(Motifs)
    score = 0
    for i in range(t):
        for j in range(k):
            if Motifs[i][j] != concensus[j]:
                score += 1
    return score
Score(["AACGTA","CCCGTT","CACCTT","GGATTA","TTCCGG"])

````
