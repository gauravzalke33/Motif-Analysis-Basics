# PROFILE MOTIFS

 we will further divide all of the elements in the count matrix by t, the number of rows in Motifs. This results in a profile matrix Profile(Motifs) for which element (i, j) is the frequency of the i-th nucleotide in the j-th column of the motif matrix (i.e., the number of occurrences of the i-th nucleotide divided by t, the number of nucleotides in the column). Note that the elements of any column of the profile matrix sum to 1.

- Sample Input:
```python
AACGTA
CCCGTT
CACCTT
GGATTA
TTCCGG
```

- Sample Output:
```python
{'A': [0.2, 0.4, 0.2, 0.0, 0.0, 0.4],'C': [0.4, 0.2, 0.8, 0.4, 0.0, 0.0],'G': [0.2, 0.2, 0.0, 0.4, 0.2, 0.2], 'T': [0.2, 0.2, 0.0, 0.2, 0.8, 0.4]}
```

````python
def Profile(Motifs):
    profile = {} # initializing the count dictionary
    k = len(Motifs[0])
    for symbol in "ACGT":
        profile[symbol] = []
        for j in range(k):
             profile[symbol].append(0)
    t = len(Motifs)
    for i in range(t):
        for j in range(k):
            symbol = Motifs[i][j]
            profile[symbol][j] += 1
    for symbol in "ACGT":
        for j in range(k):
            profile[symbol][j]=profile[symbol][j]/t
    return profile
Profile(["AACGTA","CCCGTT","CACCTT","GGATTA","TTCCGG"])
````
