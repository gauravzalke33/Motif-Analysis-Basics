# COUNT MOTIFS

To generate a count matrix from an arbitrary list of strings Motifs, we need to first initialize the count matrix, represented as a dictionary:

````python
count = {}
````

We then range over all nucleotides symbol and create a list of zeroes corresponding to count[symbol]

````python
def Count(Motifs):

    count = {}

    k = len(Motifs[0])
    for symbol in "ACGT":
        count[symbol] = []
        for j in range(k):
             count[symbol].append(0)
````

Note that the first line above sets k equal to the length of Motifs[0], the first string in Motifs, which is the length of every string in Motifs. Also, note the difference between the line count = {} (which forms an empty dictionary) and the line count[symbol] = [] (which forms an empty list). Finally, we need to range over all elements symbol = Motifs[i][j] of the count matrix and add 1 to count[symbol][j]

````python
def Count(Motifs):

    count = {}

    k = len(Motifs[0])
    for symbol in "ACGT":
        count[symbol] = []
        for j in range(k):
             count[symbol].append(0)

    t = len(Motifs)
    for i in range(t):
        for j in range(k):
            symbol = Motifs[i][j]
            count[symbol][j] += 1

    return count
````

## â“ Initial Confusion

I was confused between:
- `Motifs` being a **matrix** or a **list**.
- How `len(Motifs[0])` relates to rows or columns.
- Why we only use `Motifs[0]` to define `k` (motif length) but still access all other data.
- What `Motifs[i][j]` actually means.

---

## âœ… Clarifying Step-by-Step

### ðŸ”¹ What is `Motifs`?

`Motifs` is a **list of strings**, where each string is a **k-mer** (substring of length `k`) from a DNA string.

Example:
```python
Motifs = [
    "AGTCCG",
    "ACTTCG",
    "AGATCG"
]
```

- Each element like `"AGTCCG"` is a motif (a k-mer).
- So `Motifs[0]` is the first motif string.

---

### ðŸ”¹ Is `Motifs` a matrix?

Technically, **no**. It's a **list of strings**, but if you **imagine each string as a row**, then visually it behaves like a matrix:

| Row (i) â†“ / Col (j) â†’ | 0 | 1 | 2 | 3 | 4 | 5 |
|----------------------|---|---|---|---|---|---|
| `Motifs[0]`          | A | G | T | C | C | G |
| `Motifs[1]`          | A | C | T | T | C | G |
| `Motifs[2]`          | A | G | A | T | C | G |

---

### ðŸ”¹ What does `len(Motifs[0])` represent?

It gives the **length of the first motif** â€” i.e., the **value of `k`**.

So:
```python
k = len(Motifs[0])  # number of columns (length of each motif)
t = len(Motifs)     # number of motifs (number of rows)
```

---

### ðŸ”¹ What does `Motifs[i][j]` mean?

This accesses the **j-th character** of the **i-th motif**.

In the example:
```python
Motifs[1][2] = 'T'  # 3rd letter of the 2nd motif "ACTTCG"
Motifs[2][0] = 'A'  # 1st letter of the 3rd motif "AGATCG"
```

You can think of it like:
- `i` â†’ which **row** (which DNA string/motif)
- `j` â†’ which **column** (which position in that motif)

---

## ðŸ”š Final Insight

Now it's clear that:
- `Motifs` is a list, not a 2D matrix â€” but we can visualize it like one.
- `len(Motifs[0])` is the motif length `k`.
- `Motifs[i][j]` lets us access each nucleotide like in a matrix.



````python
def count(motifs):                    # motifs is the list which contains strings
    count = {}
    k = len(motifs[0])                # motifs[0] is the string from the list motifs 
    for symbol in "AGCT":
        count[symbol] = []
        for j in range(k):            
            count[symbol].append(0)
    
    t = len(motifs)
    for i in range(k):
        for j in range(t):
            symbol = motifs[i][j]
            count[symbol][j] += 1

    return count 
````

# ðŸ”¢ Understanding `Count(symbol)[j]` in Motif Algorithms

## ðŸ“¦ What is the `Count` matrix?

`Count` is a **dictionary of lists**, where:
- Keys are nucleotides: `'A'`, `'C'`, `'G'`, `'T'`
- Each value is a list of length `k` (motif length)  
  â†’ storing how many times that symbol appears at each position (column) across all motifs.

---

## ðŸ§  Visual Breakdown

Suppose:
```python
Motifs = [
    "AGT",
    "ACT",
    "AGG"
]
```

We want to build a `Count` dictionary from these motifs.

### Step 1: Initialize Count
```python
Count = {
    'A': [0, 0, 0],
    'C': [0, 0, 0],
    'G': [0, 0, 0],
    'T': [0, 0, 0]
}
```

### Step 2: Count each symbol column-wise

**Column 0:**
- `'A'`, `'A'`, `'A'` â†’ `A: 3`

**Column 1:**
- `'G'`, `'C'`, `'G'` â†’ `G: 2`, `C: 1`

**Column 2:**
- `'T'`, `'T'`, `'G'` â†’ `T: 2`, `G: 1`

### Final `Count`:
```python
Count = {
    'A': [3, 0, 0],
    'C': [0, 1, 0],
    'G': [0, 2, 1],
    'T': [0, 0, 2]
}
```

---

## ðŸ“Œ So What Does `Count(symbol)[j]` Mean?

- `symbol` is one of `'A'`, `'C'`, `'G'`, `'T'`
- `j` is the **column index** (position in the motif)

ðŸ‘‰ `Count('G')[1]` means:  
â†’ "How many times did the letter `'G'` appear in column `1` (i.e., second position) of all motifs?"

In our example:
```python
Count['G'][1] = 2
```

---

## ðŸ” Example Use

In consensus finding or scoring:
```python
for j in range(k):
    max_count = 0
    for symbol in "ACGT":
        if Count[symbol][j] > max_count:
            max_count = Count[symbol][j]
```
â†’ This helps find the **most frequent nucleotide at each column**.
